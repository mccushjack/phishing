```python
import numpy
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
all_sample = pd.read_csv('all_sample.csv', sep=",",header=0)
isphish = all_sample.loc[all_sample['isphish']==1]
notphish = all_sample.loc[all_sample['isphish']==0]
```

```python
sns.set_style('whitegrid')
```

```python
sns.countplot(x='isphish',data=all_sample)
```

```python
sns.swarmplot(x='isphish',y='url_len',data=all_sample,palette='Set2')
```

```python
sns.boxplot(x='isphish',y='url_len',data=all_sample,palette='rainbow')
```

```python
sns.boxplot(x='isphish',y='url_len',data=all_sample.loc[all_sample['url_len']<500],palette='rainbow')
## not phishing has lots of url >500 but excluding those Phishing tend to skew longer
```

```python
df = pd.DataFrame(all_sample)
df2 = (df[(df.url_len > 500) & (df.isphish != 0)])
```

```python
sns.countplot(x='ishttps',data=isphish)
```

```python
sns.countplot(x='ishttps',data=notphish)
```

```python
ct = pd.crosstab(df.isphish, df.ishttps,normalize='index')
stacked = ct.stack().reset_index().rename(columns={0:'value'})

sns.barplot(x=stacked.isphish, y=stacked.value, hue=stacked.ishttps)
```

```python
ct = pd.crosstab(df.isphish, df.numcoms,normalize='index')
stacked = ct.stack().reset_index().rename(columns={0:'value'})

sns.barplot(x=stacked.isphish, y=stacked.value, hue=stacked.numcoms)
```
